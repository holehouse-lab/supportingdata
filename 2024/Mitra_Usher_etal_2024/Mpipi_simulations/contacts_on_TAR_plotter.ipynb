## Code to generate 2D TAR 'structure' representation of SERF contacts
This script loads the following:
1. file with per-nucleotide chemical shift perturbation values for all residues except the three bulge nucleotides (see Figure 4B); these are outliers and are excluded from the heat map.
2. plotting coordinates that resemble the TAR stem-loop structure for CSP data
3. per-nucleotide contact frequency from simulations of TAR+SERF
4. plotting coordinates that resemble the TAR stem-loop structure for contact frequency data

Generates the left and right panels of Figure 4B.

```python
## import packages ##

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
# load file containing contact frequencies or CSPs per nucleotide #

nonbulge_csps = np.genfromtxt('TAR_CSPs+contacts_files/tar_csps_nonbulge.txt')
nonbulge_coords = np.genfromtxt('TAR_CSPs+contacts_files/csp_nonbulge_plot_coords.txt')

contacts = np.genfromtxt('TAR_CSPs+contacts_files/cg_SERF_contacts_on_TAR.txt')
contact_coords = np.genfromtxt('TAR_CSPs+contacts_files/coords_for_contacts.txt')
```

```python
## plot chemical shift perturbation data -- generates Figure 4B, right ##
# NB - masked-out bulge nucleotides added back manually with annotated values
plt.figure(num = None, figsize = (4, 6), dpi = 300, facecolor = 'w', edgecolor = 'k')

colormap = cm.get_cmap("Reds").copy()

## plot the other CSPs 
plt.scatter(nonbulge_coords[:,0], nonbulge_coords[:,1], s = 340, c = nonbulge_csps[:,1], cmap = colormap, edgecolors = 'k')

plt.xlim([-1, 2])
plt.ylim([-1, 25])

cbar = plt.colorbar()
cbar.set_label('CSP (ppm)')

## UNCOMMENT to save figure
#plt.savefig('Figure4B_TAR_NMR_CSPs.pdf', format = 'pdf', dpi = 300)

plt.show()
```

```python
## plot contact frequency data -- generates Figure 4B, left ##

plt.figure(num = None, figsize = (4, 6), dpi = 300, facecolor = 'w', edgecolor = 'k')

colormap = cm.get_cmap("Reds").copy()

## plot the other CSPs 
plt.scatter(contact_coords[:,0], contact_coords[:,1], s = 340, c = contacts, cmap = colormap, edgecolors = 'k')

plt.xlim([-1, 2])
plt.ylim([-1, 25])

cbar = plt.colorbar()
cbar.set_label('contact frequency')

## UNCOMMENT to save figure
plt.savefig('Figure4B_TAR_sim_contact_freq.pdf', format = 'pdf', dpi = 300)

plt.show()
```
