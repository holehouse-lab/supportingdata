## Code for plotting relaxation and heteronuclear NOE data from NMR measurements
This script takes five input text files that contain SERF residue numbers and measured values to generate Figures 2A, S1E, S1F, 4E, and S7E.

```python
## import packages ##

import numpy as np
import matplotlib.pyplot as plt
```

```python
## load data for plotting ##

# unbound
unbound_t1 = np.loadtxt('NMR_SERFonly_T1.txt')
unbound_t2 = np.loadtxt('NMR_SERFonly_T2.txt')
unbound_hetNOE = np.loadtxt('hetNOE_unbound.txt')

# bound
bound_t2 = np.loadtxt('HNMR_SERF-TAR_T2.txt')
bound_hetNOE = np.loadtxt('hetNOE_bound.txt')
```

```python
## plot unbound hetNOE profile -- generates Figure 2B ##

plt.figure(num = None, figsize = (5, 2), dpi = 300, facecolor = 'w', edgecolor = 'k')

plt.axhline(y = 0, lw = 0.8, color = 'k')
plt.errorbar(unbound_hetNOE[:,0], unbound_hetNOE[:,1], yerr = unbound_hetNOE[:,2], fmt = 'o', color = 'grey')

plt.xlim([1, 69])
plt.ylim([-1.5, 1.0])

plt.xlabel('residue index')
plt.ylabel('heteronuclear NOE')

# UNCOMMENT below to save figure as PDF
#plt.savefig('Figure2A_unbound_hetNOE.pdf', format = 'pdf', dpi = 300)
plt.show()
```

```python
## plot unbound T1 profile -- generates Figure S1E ##

plt.figure(num = None, figsize = (5, 2), dpi = 300, facecolor = 'w', edgecolor = 'k')

plt.scatter(unbound_t1[:,0], unbound_t1[:,1], marker = "o", color = 'grey')
plt.plot(unbound_t1[:,0], unbound_t1[:,1], color = 'k')

plt.xlim([1, 69])
plt.ylim([0, 1.2])

plt.xlabel('residue index')
plt.title('15N relaxation, T1')
plt.ylabel('T1 (1/sec)')

# UNCOMMENT below to save figure as PDF
#plt.savefig('FigureS1E_unbound_T1_profile.pdf', format = 'pdf', dpi = 300)
plt.show()
```

```python
## plot unbound T2 profile -- generates Figure S1F ##

plt.figure(figsize = (5, 2), dpi = 300, facecolor = 'w', edgecolor = 'k')

plt.scatter(unbound_t2[:,0], unbound_t2[:,1], marker = "o", color = 'grey')
plt.plot(unbound_t2[:,0], unbound_t2[:,1], color = 'k')

plt.xlim([1, 69])
plt.ylim([0, 0.8])

plt.xlabel('residue index')
plt.title('15N relaxation, T2')
plt.ylabel('T2 (1/sec)')

# UNCOMMENT below to save figure as PDF
#plt.savefig('FigureS1F_unbound_T2_profile.pdf', format = 'pdf', dpi = 300)
plt.show()
```

```python
## plot unbound hetNOE profile -- generates Figure 2B ##

plt.figure(num = None, figsize = (5, 3), dpi = 300, facecolor = 'w', edgecolor = 'k')

plt.axhline(y = 0, lw = 0.8, color = 'k')
plt.errorbar(unbound_hetNOE[:,0], unbound_hetNOE[:,1], yerr = unbound_hetNOE[:,2], fmt = 'o', color = 'grey', label = 'SERF only')
plt.errorbar(bound_hetNOE[:,0], bound_hetNOE[:,1], yerr = bound_hetNOE[:,2], fmt = 'o', color = 'forestgreen', label = 'SERF + TAR')

plt.xlim([1, 69])
plt.ylim([-1.5, 1.0])

plt.title('bound vs. unbound hetNOE')
plt.xlabel('residue index')
plt.ylabel('heteronuclear NOE')

plt.legend(loc = 'lower center')

plt.tight_layout()

# UNCOMMENT below to save figure as PDF
#plt.savefig('Figure4E_bound_unbound_hetNOE.pdf', format = 'pdf', dpi = 300)
plt.show()
```

```python
plt.figure(figsize = (5, 2), dpi = 300, facecolor = 'w', edgecolor = 'k')

plt.scatter(unbound_t2[:,0], unbound_t2[:,1], marker = "o", color = 'grey', label = 'SERF only')
plt.plot(unbound_t2[:,0], unbound_t2[:,1], color = 'grey')

plt.scatter(bound_t2[:,0], bound_t2[:,1], marker = "o", color = 'green', label = 'SERF+TAR')
plt.plot(bound_t2[:,0], bound_t2[:,1], color = 'green')
plt.errorbar(bound_t2[:,0], bound_t2[:,1], yerr = bound_t2[:,2], fmt = 'none', ecolor = 'k')

plt.xlim([1, 69])
plt.ylim([0, 1.05])

plt.xlabel('residue index')
plt.title('bound vs. unbound 15N relaxation, T2')
plt.ylabel('T2 (1/sec)')

plt.legend(loc = 'upper left')

# UNCOMMENT below to save figure as PDF
#plt.savefig('FigureS7E_bound+unbound_T2.pdf', format = 'pdf', dpi = 300)
plt.show()
```
