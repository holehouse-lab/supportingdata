## Code for plotting chemical shift perturbations on SERF with either rU30 or TAR added
This script takes two input text files that contain SERF residue numbers and chemical shift perturbation values. Generates Figures S7C-D.

```python
## import packages ##

import numpy as np
import matplotlib.pyplot as plt
```

```python
## load data for SERF + RNA chemical shift perturbations ##

serf_u30 = np.loadtxt('CSP_SERFaddU30.txt')
serf_tar = np.loadtxt('CSP_SERFaddTAR.txt')
```

```python
# plot CSPs on same y axes -- generates Figure S7C

plt.figure(figsize = (5, 2), dpi = 300, facecolor = 'w', edgecolor = 'k')

# plot SERF CSPs with added TAR
plt.scatter(serf_tar[:,0], serf_tar[:,1], color = 'green', label = '+ TAR')
plt.plot(serf_tar[:,0], serf_tar[:,1], color = 'green')

# plot SERF CSPs with added rU30
plt.scatter(serf_u30[:,0], serf_u30[:,1], marker = 's', color = 'magenta', label = '+ rU30')
plt.plot(serf_u30[:,0], serf_u30[:,1], color = 'magenta')

plt.xlim([1, 69])
plt.ylim([0, 0.85])

plt.xlabel('residue index')
plt.title('RNA-induced CSPs')
plt.ylabel('CSP (ppm)')

plt.legend(loc = 'best')

## UNCOMMENT below to save figure as PDF
#plt.savefig('FigureS7C_SERF-RNA_CSPs.pdf', format = 'pdf', dpi = 300)
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize = (5, 2), dpi = 300, facecolor = 'w', edgecolor = 'k')
ax2 = ax1.twinx()

# plot SERF CSPs with added TAR (left axis)
ax1.scatter(serf_tar[:,0], serf_tar[:,1], color = 'green', label = '+ TAR')
ax1.plot(serf_tar[:,0], serf_tar[:,1], color = 'green')

# plot SERF CSPs with added rU30 (right axis)
ax2.scatter(serf_u30[:,0], serf_u30[:,1], marker = 's', color = 'magenta', label = '+ rU30')
ax2.plot(serf_u30[:,0], serf_u30[:,1], color = 'magenta')

plt.xlim([1, 43])
ax1.set_ylim([0, 0.85])
ax2.set_ylim([0, 0.20])

ax1.set_xlabel('residue index')
plt.title('RNA-induced CSPs (rescaled)')
ax1.set_ylabel('CSP (ppm)', color = 'green')
ax1.tick_params(axis='y', labelcolor = 'green')

ax2.set_ylabel('CSP (ppm)', color = 'magenta')
ax2.tick_params(axis='y', labelcolor = 'magenta')

## UNCOMMENT below to save figure as PDF
plt.savefig('FigureS7D_SERF-RNA_CSPs.pdf', format = 'pdf', dpi = 300)
plt.show()
```
