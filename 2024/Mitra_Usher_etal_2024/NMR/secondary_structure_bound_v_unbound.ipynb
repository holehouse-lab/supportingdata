## Code for plotting secondary structure of unbound and bound SERF from NMR measurements
This script takes two input text files that contain SERF residue numbers and pre-computed ΔδCα - ΔδCβ values. Generates Figure 4D.

```python
## import packages ##

import numpy as np
import matplotlib.pyplot as plt
```

```python
## load data for plotting ##

bound = np.loadtxt('dCa_dCb_bound_serf.txt')
unbound = np.loadtxt('ncIDP_SERF_ssp.txt')
```

```python
## plot shit
plt.figure(num = None, figsize = (6, 3), dpi = 300, facecolor = 'w', edgecolor = 'k')

plt.bar(bound[:,0], bound[:,1], color = 'forestgreen', alpha = 0.7, edgecolor = 'white', linewidth = 0.1, label = 'TAR-bound')
plt.plot(unbound[:,0], unbound[:,1], color = 'dimgrey', linewidth = 2, label = 'unbound')

plt.axhline(y = 0, lw = 0.8, color = 'k')

plt.ylim([-4, 5])

plt.xlabel('residue')
plt.ylabel('ΔδCα - ΔδCβ')

plt.legend(loc = 'best')

plt.tight_layout()

# UNCOMMENT below to save figure as PDF
#plt.savefig('Figure_4D_secondary_structure.pdf', format = 'pdf', dpi = 300)
plt.show()
```
