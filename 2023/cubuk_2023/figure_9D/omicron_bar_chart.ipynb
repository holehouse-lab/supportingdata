## Omicron simulation comparison
This notebook generates the bar chart in Fig. 9D

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import matplotlib.pylab as pylab

# Set such that PDF fonts export in a manner that they
# are editable in illustrator/affinity
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
matplotlib.rcParams['legend.fontsize'] = 6

# set to define axes linewidths
matplotlib.rcParams['axes.linewidth'] = 0.5

# this defines some prefactors so inline figures look nice
# on a retina macbook. These can be commented out without any
# issue and are solely asthetic.
%matplotlib inline
%config InlineBackend.figure_format='retina'
import numpy as np
```

```python
# read in simulation data
WT      = np.loadtxt('data/NTDRBD_normalized_Ka.csv', delimiter=',')
omicron = np.loadtxt('data/Omicron_normalized_Ka.csv', delimiter=',')
P13L    = np.loadtxt('data/P13L_normalized_Ka.csv', delimiter=',')
D3133   = np.loadtxt('data/D3133_normalized_Ka.csv', delimiter=',')

WT_mean = WT[1]
WT_std = WT[2]

omicron_mean = omicron[1]
omicron_std = omicron[2]

P13L_mean = P13L[1]
P13L_std = P13L[2]

D3133_mean = D3133[1]
D3133_std = D3133[2]


```

```python
figure(None, (3.,1.8), dpi=300)
ax = plt.subplot()

plt.bar(x = ['Wuhan-Hu-1'], height = WT_mean, yerr = WT_std, capsize=3, edgecolor = "black",linewidth = .5, color='#ff0000')

plt.bar(x = ['Omicron'], height = omicron_mean, yerr = omicron_std,  capsize=3, edgecolor = "black",linewidth = .5, color='#94c9dd')

plt.bar(x = ['P13L'], height = P13L_mean, yerr = P13L_std, capsize=3, edgecolor = "black",linewidth = .5, color='#fea0a0')

plt.bar(x = [r'$\rm\Delta$31-33'], height = D3133_mean, yerr = D3133_std, capsize=3, edgecolor = "black",linewidth = .5, color='#0bace8')

ax.spines['right'].set_visible(False) 
ax.spines['top'].set_visible(False) 
#plt.yscale('log')
plt.ylabel(r'$k_A^*$ ',fontsize=8)
plt.xlabel('Variant',fontsize=8)
plt.yticks(fontsize=7)
plt.xticks(fontsize=7)
```
