```python
import numpy as np
import matplotlib

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
	
# Set such that PDF fonts export in a manner that they
# are editable in illustrator/affinity
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

# set to define axes linewidths
matplotlib.rcParams['axes.linewidth'] = 0.5

# this defines some prefactors so inline figures look nice
# on a retina macbook. These can be commented out without any
# issue and are solely asthetic.
%matplotlib inline
%config InlineBackend.figure_format='retina'

# UPDATE 2020-12-31 (my preferred font is Avenir...)
font = {'family' : 'avenir',
    	'weight' : 'normal'}

matplotlib.rc('font', **font)
```

```python
ev_data = {}
lengths = list(range(50,550,50))
for l in lengths:
    ev_data[l] = np.loadtxt(f'data/ev_a_{l}.cvs')
```

```python
frc_data = {}
lengths_frc = [51,101,151,251,351]
for l in lengths_frc:
    frc_data[l] = np.loadtxt(f'data/frc_a_{l}.cvs')
```

```python
figure(None, (3,2), dpi=350)
for l in lengths:
    plt.plot(np.arange(1,len(ev_data[l])+1), ev_data[l], linewidth=0.7, alpha=0.2, color='g')
    plt.plot(len(ev_data[l]),ev_data[l][-1],'.', ms=2, color='g')


for l in lengths_frc:
    plt.plot(np.arange(1,len(frc_data[l])+1), frc_data[l], linewidth=0.7, alpha=0.2, color='k')
    plt.plot(len(frc_data[l]),frc_data[l][-1],'.', ms=2, color='k')
    
    
plt.xlim([0,550])    
plt.ylim([0,250])    

plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.ylabel(r'$\langle \langle r_{(i,j)} \rangle \rangle$  ($\rm\AA$)', fontsize=8)
plt.xlabel(r'$|i-j|$ (residues)', fontsize=8)

plt.tight_layout()
plt.savefig('panel_1E.pdf')
```
