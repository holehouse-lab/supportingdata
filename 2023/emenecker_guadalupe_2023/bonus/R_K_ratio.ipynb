```python
import protfasta
from sparrow import Protein
```

```python
seqs = protfasta.read_fasta('../sequence_analysis/synthetic_idrs.fasta')
```

```python
for s in seqs:
    p = Protein(seqs[s])
    
    print(f'On sequence {s}')
    print(f"Fraction K: {p.amino_acid_fractions['K']}")
    print(f"Fraction R: {p.amino_acid_fractions['R']}")
    print('-----------')
```
