```python
from localcider.sequenceParameters import SequenceParameters
from localcider import plots
```

```python
import protfasta
```

```python
s = protfasta.read_fasta('synthetic_idrs.fasta')
```

```python
all_seqs = []
labels = []
for x in s:
    all_seqs.append(SequenceParameters(s[x]))
    labels.append(f"{x}")
    
    print(f"{x} - {SequenceParameters(s[x]).get_phasePlotRegion()} - fneg: {SequenceParameters(s[x]).get_fraction_negative()} ")
```

```python
plots.show_multiple_phasePlot2(all_seqs)

plots.save_multiple_phasePlot2(all_seqs, 'das_pappu.pdf', saveFormat='pdf')
```
