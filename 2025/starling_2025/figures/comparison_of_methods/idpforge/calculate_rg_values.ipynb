## About
This notebook will calculate $R_g$ and $R_e$ values for ensembles generated by IDPForge.

OK so... for some reason, IDPForge does not generate PDB files with a consistent topology. I don't know what fresh hell would lead to this bug, but it means you can't concatanate the PDB files together into a single PDB/XTC/MDTraj ensemble because of this inconsistent toplogy.

So, this code actually sequentially reads in each PDB file and for EACH PDB file calculates the Rg and Re values for each PDB file, and then writes that out to a file called `rg_values.csv` and `re_values.csv`, which is fine, but VERY slow.

```python
from tqdm import tqdm
import pandas as pd
import os
import numpy as np
```

```python
df = pd.read_csv('data/all_comparison_data_WITH_STARTING.csv')
names = df['name'].to_list()
```

## WARNING
The cell below takes ~7 hours to run...

```python
count = 400
for n in tqdm(names):

    # if the directory for sequence n in is present
    if os.path.isdir(f"predictions/400_confs/{n}/"):

        # and if we DON'T already have an rg_values.csv file
        if not os.path.isdir(f"predictions/400_confs/{n}/rg_values.csv"):
    
            rg_values = []
            re_values = []
            for i in range(1, count+1):
                pdb_file = f"predictions/400_confs/{n}/{i}_relaxed.pdb"
                p = SSTrajectory(pdb_file,pdb_file).proteinTrajectoryList[0]
                rg_values.append(p.get_radius_of_gyration()[0])
                re_values.append(p.get_end_to_end_distance()[0])
    
            np.savetxt(f'predictions/400_confs/{n}/rg_values.csv', rg_values)
            np.savetxt(f'predictions/400_confs/{n}/re_values.csv', re_values)
            

    
```
