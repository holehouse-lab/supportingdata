```python
import numpy as np
import matplotlib

import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
	
# Set such that PDF fonts export in a manner that they
# are editable in illustrator/affinity
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

# set to define axes linewidths
matplotlib.rcParams['axes.linewidth'] = 0.5

# this defines some prefactors so inline figures look nice
# on a retina macbook. These can be commented out without any
# issue and are solely asthetic.
%matplotlib inline
%config InlineBackend.figure_format='retina'

# UPDATE 2020-12-31 (my preferred font is Avenir...)
font = {'family' : 'avenir',
    	'weight' : 'normal'}

matplotlib.rc('font', **font)

```

```python
from finches import  CALVADOS_frontend, Mpipi_frontend
from sparrow import Protein

mf = Mpipi_frontend()
cf = CALVADOS_frontend()
```

```python

# note the [::-1] which reverses the order so the sequences descend from highes-to-lowest aromatic fraction
# when plotting
# order here is
# minus minus, minus, wt, plus
a1_seqs = ['GSMASASSSQRGRSGSGNSGGGRGGGFGGNDNSGRGGNSSGRGGFGGSRGGGGSGGSGDGYNGSGNDGSNSGGGGSSNDFGNSNNQSSNSGPMKGGNFGGRSSGGSGGGGQYSAKPRNQGGSGGSSSSSSSGSGRRS',
'GSMASASSSQRGRSGSGNSGGGRGGGFGGNDNFGRGGNSSGRGGFGGSRGGGGYGGSGDGYNGFGNDGSNSGGGGSSNDFGNYNNQSSNFGPMKGGNFGGRSSGGSGGGGQYSAKPRNQGGYGGSSSSSSSGSGRRF',
'GSMASASSSQRGRSGSGNFGGGRGGGFGGNDNFGRGGNFSGRGGFGGSRGGGGYGGSGDGYNGFGNDGSNFGGGGSYNDFGNYNNQSSNFGPMKGGNFGGRSSGGSGGGGQYFAKPRNQGGYGGSSSSSSYGSGRRF',
'GSMAFASSFQRGRYGSGNFGGGRGGGFGGNDNFGRGGNFSGRGGFGGSRGGGGYGGSGDGYNGFGNDGSNFGGGGSYNDFGNYNNQSSNFGPMKGGNFGGRSSGGSYGGGQYFAKPRNQGGYGGSSFSSSYGSGRRF',][::-1]



figure(num=None, figsize=(3.3, 2.21), dpi=300, facecolor='w', edgecolor='k')
ax = plt.gca()

colors = {}
colors[0] = '#70cbce'
colors[1] = '#231f20'
colors[2] = '#559ccf'
colors[3] = '#274fa2'


names = {}
names[0] = '20% aro'
names[1] = '15% aro'
names[2] = '10% aro'
names[3] = '5% aro'


wt_base = max(mf.build_phase_diagram(a1_seqs[1])[3])

for idx, d in enumerate(a1_seqs):

    XOUT = mf.build_phase_diagram(d)

    plt.plot(XOUT[0], XOUT[3]/wt_base, '-', color=colors[idx],linewidth=0.8)
    plt.plot(XOUT[1], XOUT[3]/wt_base,'-', color=colors[idx],linewidth=0.8, label=names[idx])    
    plt.plot(XOUT[2][0], XOUT[2][1]/wt_base,'.', color=colors[idx],linewidth=0.5, ms=8,  mew=0)    
    

#plt.legend(fontsize=7, frameon=False)

plt.xlim ([0.00001,0.7])

## uncomment to show as log slcae
#plt.xscale('log')

plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.xscale('log')
plt.tight_layout()
plt.xlabel(r'Concentration ($\phi$)', fontsize=8)
plt.ylabel('Temperature\n(Arbitrary units)', fontsize=8)
plt.tight_layout()
plt.savefig('out_figures/A1_phase_diagram.pdf')


```
