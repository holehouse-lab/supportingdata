```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np

from finches import Mpipi_frontend

# laf-1 RGG sequence
rgg = 'MESNQSNNGGSGNAALNRGGRYVPPHLRGGDGGAAAAASAGGDDRRGGAGGGGYRRGGGNSGGGGGGGYDRGYNDNRDDRDNRGGSGGYGRDRNYEDRGYNGGGGGGGNRGYNNNRGGGGGGYNRQDRGDGGSSNFSRGGYNNRDEGSDNRGSGRSYNNDRRDNGGDG'

```

```python
eps = []
for s in [0.125, 0.250, 0.3]:
    mf = Mpipi_frontend(salt=s)
    eps.append(mf.epsilon(rgg,rgg))
```

```python
# print eps at salt=0
mf = Mpipi_frontend(salt=0)
mf.epsilon(rgg,rgg)
```

```python
# the B2 values here come straight from Wei et al. How do I know? I am a co-1st author of Wei et al., so have all
# the original data! These are the data ploted as Fig 2c.
data = np.array([-30.5, -20.3, -17.9])
```

```python
# from wei et al
figure(None, (1.8, 1.5), dpi=450)

plt.plot(eps, [-30.5, -20.3, -17.9],'or', mew=0.5, mec='k')
fit_data = np.polyfit(eps, data, 1)

xmin = -8.4
xmax = -9.0
xdata = np.arange(xmax, xmin, 0.01)
plt.xlim([xmax, xmin])
plt.ylim([-32, -14])
plt.plot(xdata, xdata*fit_data[0] + fit_data[1], '--r', linewidth=0.5, zorder=-20)
plt.xlabel(r'$\epsilon$ [prediction]', fontsize=8)
plt.ylabel(r'$B_{2}$ (L mol kg$^{-2}$) [experiment]', fontsize=8)
plt.xticks(fontsize=7)
plt.yticks(fontsize=7)

plt.tight_layout()
plt.savefig('rgg_b2.pdf')
```
